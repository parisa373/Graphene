```python
import matplotlib.pyplot as plt
import numpy as np

a = 1

def f(x, y):
    return np.sqrt(3 + 2*(np.cos(0.5*x*np.sqrt(3)))*np.cos(y*a*0.5) - 2*np.sin(0.5*x*np.sqrt(3))*np.sin(y*a*0.5) + 2*np.cos(2*a*y) + 2*(np.exp(1j*0.5*a*x*np.sqrt(3)))*np.cos(y*a*0.5))

x = np.linspace(-4,4, 500)
y = np.linspace(-4,4, 500)

X, Y = np.meshgrid(x, y)
Z1 = f(X, Y)
Z2 = -1*f(X, Y)

#ax = plt.axes(projection='3d')
#ax.plot_surface(X, Y, Z1, cmap='viridis', edgecolor='blue')
#ax.plot_surface(X, Y, Z2, cmap='viridis', edgecolor='red')


fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z1, 50, cmap='viridis', color='red')
ax.contour3D(X, Y, Z2, 50, cmap='inferno', color='blue')
ax.set_xlabel(r'$K_x$')
ax.set_ylabel(r'$K_y$')
ax.set_zlabel(r'$E_k/\gamma$');

fig.set_figwidth(7)
fig.set_figheight(7)
```

```python
Z3 = f(x, y)
Z4 = -1*f(x, y)
plt.ylabel(r'$E_{k}/\gamma$')
plt.xlabel(r'$K_x~&~K_y$')
plt.grid()
plt.plot(x, Z3)
plt.plot(x, Z4)
```
